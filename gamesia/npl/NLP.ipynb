# CountVectorizer

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

df = pd.read_csv('../all_games.csv', encoding='utf-8')
df.head()
```

```python
summary = df['summary'][0]
#summary = list(summary)
#3summary
#summary
summary = ['As a young boy, Link is tricked by Ganondorf, the King of the Gerudo Thieves. The evil human uses Link to gain access to the Sacred Realm, where he places his tainted hands on Triforce and transforms the beautiful Hyrulean landscape into a barren wasteland. Link is determined to fix the problems he helped to create, so with the help of Rauru he travels through time gathering the powers of the Seven Sages.']
```

```python
#summary = df['summary'].values.astype('U')
#df['summary'].values.tolist()

#df['sy'].apply(lambda x: np.str_(x))
summary = df['summary'].apply(lambda x: np.str_(x))
summary
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer()
count_matrix = count_vect.fit_transform(summary)
count_array = count_matrix.toarray()

summaryData = pd.DataFrame(data=count_array, columns=count_vect.get_feature_names())
summaryData.head()
```

```python
summaryData.columns
```

```python
vectorizer = CountVectorizer()

vectorizer.fit(summary)
print("Vocabulary: ", vectorizer.vocabulary_)

vector = vectorizer.transform(summary)

print("Encoded Document is:")
print(vector.toarray())
```
