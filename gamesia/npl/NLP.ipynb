# CountVectorizer

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

df = pd.read_csv('../all_games.csv', encoding='utf-8')
df.head()
```

```python
#summary = df['summary'].values.astype('U')
#df['summary'].values.tolist()

#df['sy'].apply(lambda x: np.str_(x))
summary = df['summary'].apply(lambda x: np.str_(x))
summary
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer()
count_matrix = count_vect.fit_transform(summary)
count_array = count_matrix.toarray()

summaryData = pd.DataFrame(data=count_array, columns=count_vect.get_feature_names())
summaryData.head()
```

```python
summaryData.columns
```

```python
vectorizer = CountVectorizer()

vectorizer.fit(summary)
print("Vocabulary: ", vectorizer.vocabulary_)

vector = vectorizer.transform(summary)

print("Encoded Document is:")
print(vector.toarray())
```

```python
vector.shape
```
